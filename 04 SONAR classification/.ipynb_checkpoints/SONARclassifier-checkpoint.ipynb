## The task is to train a model to discriminate between SONAR signals bounced off a metal cylinder and those bounced off a roughly cylindrical rock. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('sonar_dataset.csv', header=None)
df.head(10)
```

```python
df.describe()
```

```python
df[60].value_counts()
```

#### The dataset is pretty much ready for us already so we can use it as it is to generate features and labels. 

```python
X = df.drop(60, axis=1)
y = df[60]
y.head()
```

```python
y = pd.get_dummies(y, drop_first=True)
y.replace({True: 1, False:0}, inplace=True)
y.head()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_test.shape
```

### Building a model

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from keras import Sequential
from keras.layers import Dense, Dropout
```

```python
# Avoid OOM errors by setting GPU Memory Consumption Growth
gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus: 
    tf.config.experimental.set_memory_growth(gpu, True)
```

```python
tf.config.list_physical_devices('GPU')
```

```python
model = Sequential([
    Dense(units=50, input_shape=(60,), activation='relu'),
    Dense(units=25, activation='relu'),
    Dense(units=10, activation='relu'),
    Dense(units=1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
logdir = 'logs'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(X_train, y_train, epochs=50, batch_size=8, validation_data=(X_test, y_test), callbacks=[tensorboard_callback])
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
#plt.ylim([0.35, 0.6])
plt.show()
```

### We can see that this model is overfitting the data. So let us simplify the model

```python
model2 = Sequential([
    Dense(units=50, input_shape=(60,), activation='relu'),
    Dropout(0.5),
    Dense(units=25, activation='relu'),
    Dropout(0.2),
    Dense(units=10, activation='relu'),
    #Dropout(0.5),
    Dense(units=1, activation='sigmoid')
])
```

```python
model2.summary()
```

```python
model2.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
logdir = 'logs2'
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model2.fit(X_train, y_train, epochs=50, batch_size=8, validation_data=(X_test, y_test), callbacks=[tensorboard_callback])
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

#### We can see that with this new model, the performance is much better

```python
predictions = model.predict(X_test)
y_pred = []
for element in predictions:
    if element>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)
import seaborn as sn
plt.figure(figsize = (6,5))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
