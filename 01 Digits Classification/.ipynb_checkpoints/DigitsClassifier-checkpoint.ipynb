```python
import tensorflow as tf
```

```python
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Plotting data
num_row = 4
num_col = 5
# plot images
fig, axes = plt.subplots(num_row, num_col, figsize=(1.5*num_col,2*num_row))
for i in range(20):
    ax = axes[i//num_col, i%num_col]
    ax.imshow(X_train[i], cmap='gray')
    ax.set_title('Label: {}'.format(y_train[i]))
plt.tight_layout()
plt.show()
```

```python
X_train_flat = X_train.reshape(len(X_train), 28*28) / 255.
X_test_flat = X_test.reshape(len(X_test), 28*28) / 255.
print("X_train flattened: ", X_train_flat.shape)
print("X_test flattened: ", X_test_flat.shape)
```

<b> Creating the model </b>

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
gpus = tf.config.list_physical_devices('GPU')
if gpus:
  try:
    # Currently, memory growth needs to be the same across GPUs
    for gpu in gpus:
      tf.config.experimental.set_memory_growth(gpu, True)
    logical_gpus = tf.config.list_logical_devices('GPU')
    print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
  except RuntimeError as e:
    # Memory growth must be set before GPUs have been initialized
    print(e)
```

#### TensorFloat-32, or TF32 for short, is a math mode for NVIDIA Ampere GPUs and above. TensorFloat-32 execution causes certain float32 ops, such as matrix multiplications and convolutions, to run much faster on such GPUs but with reduced precision. This reduced precision should not impact convergence of deep learning models in practice.

#### TensorFloat-32 is enabled by default. TensorFloat-32 is only supported on NVIDIA GPUs starting with the Ampere generation, so older NVIDIA GPUs and other hardware will use the full float32 precision regardless of whether TensorFloat-32 is enabled or not. 

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.optimizers import Adam

tf.config.experimental.enable_tensor_float_32_execution(False)

model = Sequential([
    Dense(units=50, activation='relu', input_shape=(784,)),
    Dense(units=30, activation='sigmoid'),
    Dense(units=10, activation='softmax')
])

model.compile(
    loss = SparseCategoricalCrossentropy(), 
    optimizer=Adam(learning_rate = 1e-3), 
    metrics=['accuracy']
)

model.fit(X_train_flat, y_train, epochs = 10)
```

### Now let us evaluate it on a test dataset

```python
model.evaluate(X_test_flat, y_test)
```

#### Let us plot our predictions and check if they are accurate

```python
num_row = 6
num_col = 6

y_pred = model.predict(X_test_flat)
# plot images
fig, axes = plt.subplots(num_row, num_col, figsize=(1.5*num_col,2*num_row))
for i in range(36):
    ax = axes[i//num_col, i%num_col]
    ax.imshow(X_test[i], cmap='gray')
    ax.set_title('Predicted: {}'.format(np.argmax(y_pred[i])))
plt.tight_layout()
plt.show()
```

#### Plotting Confusion matrix

```python
predicted_labels = [np.argmax(i) for i in y_pred]
cm = tf.math.confusion_matrix(labels=y_test, predictions=predicted_labels)
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
